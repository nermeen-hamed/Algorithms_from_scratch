# Apriori_Algorithm

```python
#read_lines=list(line.strip().split(',')) did not work

import csv
def load_transaction(path_file):
    Transactions={}
    with open(path_file,'r')as f:
        read_lines = csv.reader(f, delimiter=',')
        count = 0
        for line in read_lines:
            count+=1
            unique_line=list(np.unique(line))
            Transactions[count]=set(unique_line)
    return(Transactions)
```

```python
t=load_transaction('newUpdated.csv')
```

```python
t
```

```python
def frequence(items_lst, transactions,min_supp):
    items_counts ={}
    for i in items_lst:
        subset ={i}
        for j in transactions.items():
            if (subset.issubset(set(j[1]))):
                if i in items_counts:
                    items_counts[i] += 1
                else:
                    items_counts[i] = 1
    if '' in items_counts:
        items_counts.pop('')
    for key,value in list(items_counts.items()):
        if(value<min_supp):
            del items_counts[key]
    return items_counts
```

```python
items_lst=set()
for i in t.values():
    for j in i:
        items_lst.add(j)
```

```python
item_count=frequence(items_lst, t,15)
```

```python
def freq_(l,t,min_supp):
    itemSets={}
    for i in l:
        subset =frozenset(i)
        for j in t.items():
            if(subset.issubset(set(j[1]))):
                if frozenset(i) in itemSets:
                    itemSets[frozenset(i)] += 1
                else:
                    itemSets[frozenset(i)] = 1
    for key,value in list(itemSets.items()):
        if(value<min_supp):
            del itemSets[key]
    return itemSets 
```

```python
def subsets(itemSets):
    final_set=[]
    if(isinstance(itemSets, dict)):
        items=list(itemSets.keys())
        for i in range(len(items)):
            for j in range((i+1), len(items)):
                itemset_i ={items[i]}
                itemset_j ={items[j]}
                union_sets = itemset_i.union(itemset_j)
                if union_sets not in final_set:
                    final_set.append(union_sets)
    else:
        for i in range(len(itemSets)):
            for j in range((i+1), len(itemSets)):
                itemset_i =itemSets[i]
                itemset_j =itemSets[j]
                union_sets = itemset_i.union(itemset_j)
                if union_sets not in final_set:
                    final_set.append(union_sets)
    return final_set;
```

```python
def support(frq_items,t):
    item_with_support={}
    for key,value in frq_items.items():
        support=value/len(t)
        item_with_support[key]=support
    return item_with_support
```

```python
from itertools import combinations
def get_confidence(items_sets,t):
    list1=[]
    list2=[]
    for value in items_sets:
        compining_value=[frozenset(i) for i in combinations(value,len(value)-1)]
        for A in compining_value:
            
            B=value-A
            AB=value
            support_of_AB=0
            support_of_A=0
            support_of_B=0
            for s in t.items():
                if(A.issubset(s[1])):
                    support_of_A+=1
                if(B.issubset(s[1])):
                    support_of_B+=1
                if(AB.issubset(s[1])):
                    support_of_AB+=1
            temp = support_of_AB/support_of_A*100
            print(str(list(A))+" -> "+str(list(B))+" = "+str(temp)+"%")
            temp = support_of_AB/support_of_B*100
            print(str(list(B))+" -> "+str(list(A))+" = "+str(temp)+"%")
            print('\n')
            break
```
